```python
def merge_sort(l, start, end):
    print(start, end)
    # If list is empty or has one element, return
    if len(l[start:end]) <= 1:
        return
    mid = (end-start)//2
    merge_sort(l, start, mid)
    merge_sort(l, mid+1, end)
    merge(l, start, mid, end)
```

```python
def merge(l, start, mid, end):
    c = []
    a = l[start:mid+1]
    b = l[mid+1:end+1]
    print(a,b)
    i = 0
    j = 0
    # print(start, end)
    while i != len(a) and j != len(b):
        print(i,j)
        if a[i] < b[j]:
            c.append(a[i])
            i = i+1
        elif a[i] > b[j]:
            c.append(b[j])
            j = j+1
        print(c)
        
    if i == len(a):
        while j != len(b):
            c.append(b[j])
            j = j+1
            
    if j == len(b):
        while i != len(a):
            c.append(a[i])
            i = i+1
            
    l[start:end+1] = c
```
