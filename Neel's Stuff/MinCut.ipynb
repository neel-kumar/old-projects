```python
import numpy as np
```

```python
def rand_edge(graph):
    x = y = 0
    x = np.random.randint(0, len(graph))
    y = np.random.randint(1, len(graph[x]))
    #x = 1
    #y = 1
    return x,y

# Remove all repetitions of first items except the first one
def remove_self_loop(a):
    b = [a[0]]
    for i in range(1, len(a)):
        if a[i] != a[0]:
            b.append(a[i])
    return b

def index(graph,x,y):
    y_value = graph[x][y]
    for i in range(0, len(graph)):
        if graph[i][0] == y_value:
            return i

def collapse_edge(graph,x,y):
    x_value = graph[x][0]
    y_value = graph[x][y]
    y_index = index(graph,x,y)
    #print(graph[y])
    for i in range(1, len(graph[y_index])):
        graph[x].append(graph[y_index][i])
        #print(i, graph[y_value-1])
    for i in range(0, len(graph)):
        for j in range(1, len(graph[i])):
            if graph[i][j] == y_value:
                graph[i][j] = x_value
    #print(x, graph)
    #print("Before remove_self_loop", graph)
    graph[x] = remove_self_loop(graph[x])
    #print("After remove_self_loop", graph)
    graph.remove(graph[y_index])
    #print("After remove node", graph)
```

```python
def min_cut(graph):
    #print(len(graph), "len")
    if len(graph) == 2:
        #print(len(graph[0]))
        #print(graph)
        #print("min_cut")
        return len(graph[0])-1
    x,y = rand_edge(graph)
    #print("edge:",x,y)
    collapse_edge(graph,x,y)
    min_cut(graph)
    return len(graph[0])-1
```

```python
print("[[1,2,4], [2,1,3,4], [3,2,4], [4,1,2,3]]")
min_cut([[1,2,4], [2,1,3,4], [3,2,4], [4,1,2,3]])
```

```python
least = []
for i in range(1000):
    cut = min_cut([[1,2,4], [2,1,3,4], [3,2,4], [4,1,2,3]])
    least.append(cut)
least.sort()
print(least)
print("min cut",least[0])
```

```python
def read_file():
    filename = "/Users/soumitra/Desktop/min_cut.txt"
    file = open(filename, 'r')
    l1 = file.readlines()
    l = []
    for i in l1:
        l2 = i.split()
        for j in range(0, len(l2)):
            l2[j] = int(l2[j])
        l.append(l2)
    return l
```
