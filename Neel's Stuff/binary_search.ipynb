```python
def binary_search(input_array, value):
    i = 0
    j = -1
    while i != j:
        middle = len(input_array[i:j])//2
        if middle < value:
            i = middle+1
        elif middle > value:
            j = middle
        else:
            return middle
    if input_array[i] == value:
        return i
    return -1

test_list = [1,3,9,11,15,19,29]
test_val1 = 25
test_val2 = 15
print(binary_search(test_list, test_val1))
print(binary_search(test_list, test_val2))
```

```python
# [i, j]
def binary_search(input_array, value, i, j):
    if i > j:
        return -1
    middle = ((j-i)//2)+i
    if input_array[middle] < value:
        return binary_search(input_array, value, middle+1, j)
    elif input_array[middle] > value:
        return binary_search(input_array, value, i, middle-1)
    else:
        return middle
    return -1

test_list = [1,3,9,11,15,19,29]
test_val1 = 25
test_val2 = 15
print(binary_search(test_list, test_val1, 0, 6))
print(binary_search(test_list, test_val2, 0, 6))
```
